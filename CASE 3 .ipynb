```r
library(tidyverse)
library(tidyr)
library(rvest)
library(dplyr)
library(janitor)
library(ggplot2)
library(scales)
library(matlab)

#Leser inn html fila

Motor <- "https://www.motor.no/aktuelt/motors-store-vintertest-av-rekkevidde-pa-elbiler/217132"%>%
  read_html() %>%
  html_nodes(xpath ='/html/body/article/section/div[4]/div[4]/div[1]/table') %>%
  html_table(header= 1, fill = TRUE) %>%
  as.data.frame()

Motor1 <- Motor [-c(19, 26), ]


#Rydder opp i kolonnene. Fjerner WLTP tall, km og x. Lager egen kolonne med rekkevidde, og kWh

Motor1 <- Motor %>%
  filter(!str_detect(Avvik, "x")) %>%
  separate(`WLTP.tall`, sep = "/", into=c("Rekkevidde", "KWH")) %>%
  mutate(STOPP = as.numeric(gsub("km", "", STOPP))) %>%
  mutate(Rekkevidde = as.numeric(gsub("km", "", Rekkevidde)))
view (Motor1)


#Plotter som graf, og alle punktene er under 45 graders linje

Motor1 %>%
  ggplot(aes(x = Rekkevidde, y = STOPP)) +
  geom_point() +
  geom_abline(color = "red") +
  labs(
    title = "Rekkevidde tapt i kulden",
    y = "STOPP",
    x = "WLTP")

#Oppgave 2


#Benytter lm funksjonen å ser at det nå er flere bilmerker som krysser 45 graders linja


RV <- lm(Rekkevidde ~ STOPP, data = Motor1)

print(Motor1)


Motor1 %>%
  ggplot(aes(x = Rekkevidde, y = STOPP)) +
  geom_point() +
  geom_smooth(method = lm) +
  labs(
    title = "",
    y = "STOPP",
    x = "WLTP")
 

```
